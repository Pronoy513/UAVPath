```python
#mask generator.
import cv2
import numpy as np

def create_img(image):
  image = cv2.resize(image, (720, 720))
  output_path = "x_output.jpg"
  cv2.imwrite(output_path, image)
  return output_path

def create_color_mask(image):


  # Optional resizing (adjust new_size as needed)
  image = cv2.resize(image, (720, 720))

  # Convert to BGR (if needed)
  image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # Assuming your image is in RGB format

  # Define color ranges (adjusted based on your input)
  lower1 = np.array([250, 0, 0], dtype="uint8")  # Red: Lower bound (adjust as needed)
  upper1 = np.array([255, 50, 50], dtype="uint8")  # Red: Upper bound (adjust as needed)
  lower2 = np.array([0, 200, 200], dtype="uint8")  # Cyan: Lower bound (adjust as needed)
  upper2 = np.array([50, 255, 255], dtype="uint8")  # Cyan: Upper bound (adjust as needed)

  # Create masks for each color
  mask1 = cv2.inRange(image, lower1, upper1)
  mask2 = cv2.inRange(image, lower2, upper2)


  # Combine masks with colors
  color_image = image.copy()  # Copy the image for coloring the mask
  color_image[mask2 > 0] = 1  # Paint cyan areas (BGR)
  color_image[mask1 > 0] = 2  # Paint red areas (BGR)

  # Set background to black
  color_image[np.where((mask1 == 0) & (mask2 == 0))] = 0  # Set background to black (BGR)

  # Save the image with colored mask
  output_path = "c_output.jpg"
  color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
  cv2.imwrite(output_path, color_image)
  analyze_image(output_path)
  return output_path

# # Example usage
# image_path = "xyz.jpg"
# image = cv2.imread(image_path)  # Read your image
# output_path = create_color_mask(image)  # Generate and save the colored mask

# print(f"Colored mask saved at: {output_path}")
```

```python
image_path="/content/drive/MyDrive/Dataset/maskfinal/109.png"
from PIL import Image
import numpy as np

def analyze_image(image_path):
  """
  This function reads an image, converts it to grayscale, converts it to a NumPy array,
  and prints the number of occurrences of each unique pixel value.

  Args:
      image_path (str): Path to the image file.
  """
  try:
    # Open the image in grayscale mode
    img = Image.open(image_path).convert('L')
  except FileNotFoundError:
    print(f"Error: Image file not found at {image_path}")
    return

  # Convert the image data to a NumPy array
  image_array = np.array(img)

  # Find the unique values and their counts
  unique_values, counts = np.unique(image_array, return_counts=True)

  # Print the results
  print("Unique pixel values and their occurrences:")
  for value, count in zip(unique_values, counts):
    print(f"Value {value}: {count} occurrences")



```

```python
#Main Driver to access the files and create the required datasets with appropriate processing and naming.
import os
import cv2
import numpy as np

# Replace with actual paths to your directories
dir1 = "/content/drive/MyDrive/Dataset/pre-img"
dir2 = "/content/drive/MyDrive/Dataset/pre-mask"
dir3 = "/content/drive/MyDrive/Dataset/imgfinalx"
dir4 = "/content/drive/MyDrive/Dataset/maskfinalx"
os.makedirs(dir3, exist_ok=True)
os.makedirs(dir4, exist_ok=True)

# Track processed filenames (set) and counter (int)
processed_filenames = set()
counter = 0

# Get all image files in dir1
image_files1 = []
for filename in os.listdir(dir1):
    if filename.endswith(".png"):  # Adjust for other image formats if needed
        image_files1.append(os.path.join(dir1, filename))

for image_file1 in image_files1:
    # Extract base filename
    base_filename = os.path.splitext(os.path.basename(image_file1))[0]

    # Check if already processed
    if base_filename not in processed_filenames:
        processed_filenames.add(base_filename)  # Add to processed set

        # Process image from dir1
        img1 = cv2.imread(image_file1)
        processed_img1 = create_img(img1)  # Call your create_img function

        # Create new filename for processed img1
        counter += 1
        new_filename1 = f"{counter:03d}.png"  # Adjust extension if needed

        # Save processed img1 to dir3
        cv2.imwrite(os.path.join(dir3, new_filename1), cv2.imread(processed_img1))

        # Process mask from dir2 with the same base filename
        mask_file2 = os.path.join(dir2, base_filename + ".png")  # Assume same extension
        if os.path.exists(mask_file2):  # Check if mask exists
            img2 = cv2.imread(mask_file2)
            processed_mask2 = create_color_mask(img2)  # Call your create_color_mask function

            # Create new filename for processed mask2
            new_filename2 = f"{counter:03d}.png"  # Adjust extension if needed

            # Save processed mask2 to dir4
            cv2.imwrite(os.path.join(dir4, new_filename2), cv2.imread(processed_mask2))

        print(f"Processed image {counter}: {new_filename1} and {new_filename2}")

print(f"Total processed images: {counter}")
```

```python
import cv2
import numpy as np

def visualize_grayscale_mapping(input_path, output_path):
    """
    Reads a grayscale image, maps each pixel value to a specific color, and saves the colored image.

    Args:
        input_path (str): Path to the grayscale image file.
        output_path (str): Path to save the colored image.
    """

    img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale

    # Create a dictionary for color mapping
    color_map = {
        0: (0, 0, 0),    # Black
        1: (0, 0, 255),  # Red
        2: (255, 0, 0),  # Green
        3: (255, 0, 0),  # Blue
        4: (255, 0, 255) # Purple
    }

    # Create a new image to hold the colored result
    colored_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

    # Loop through each pixel and assign color based on value
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            pixel_value = img[y, x]
            colored_img[y, x] = color_map.get(pixel_value, (255, 255, 255))  # Assign color from the map, default to white

    # Save the colored image
    cv2.imwrite(output_path, colored_img)

# Example usage
input_path = "/content/drive/MyDrive/Dataset/maskfinal/109.png"  # Replace with your image path
output_path = "output.jpg"  # Replace with your desired output path
visualize_grayscale_mapping(input_path, output_path)
```
